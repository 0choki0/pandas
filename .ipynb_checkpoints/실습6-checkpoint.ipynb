```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
i_groups = iris.groupby(iris.species)
```

```python
i_groups
```

```python
i_groups.mean()
```

```python
pd.DataFrame(i_groups).loc[1].values[1]['sepal_width'].min()
```

```python
def peak_to_peak_ratio(x):
    return x.max()/x.min()
```

```python
peak_to_peak_ratio(iris.sepal_length)
```

```python
i_groups.agg(peak_to_peak_ratio)
```

```python
def top3_petal_length(df):

    return df.sort_values(by='petal_length', ascending=False)[:3]
```

```python
top3_petal_length(iris)
```

```python
i_groups.apply(top3_petal_length)
```
